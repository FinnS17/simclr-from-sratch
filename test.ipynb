```python
import random

selected_images = [x for x in random.sample(range(1000), 5)]

fig, axes = plt.subplots(nrows=len(selected_images), ncols=2, figsize=(6,12))

for i, idx in enumerate(selected_images):
    clean = dataset[idx][0]
    changed = transform(dataset[idx][0])

    axes[i][0].imshow(clean)
    axes[i][1].imshow(to_display_img(changed))
plt.tight_layout()

```
